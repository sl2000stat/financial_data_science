## B.1 GARCH(m,s) Model

**Heteroscedastic  return decomposition:**

$$
r_t = \mu_{t-1} + \epsilon_t; \quad \epsilon_t \equiv \sigma_{t-1} \, \times \, \eta_t; \; \eta_t \sim iid(0,1).
$$

with

$$
\\
$$

$$
\sigma^2_{t-1} \overbrace{:=}^{GARCH(m,s)} \alpha_0 + \sum_{i=1}^m \alpha_i \, \times \, \epsilon^2_{t-i} \, + \, \sum_{j=1}^s \beta_j \, \times \, \sigma^2_{t-1-j}.
$$

with $\alpha_0 > 0$, $\alpha_i \geq 0, \forall i \in [1,...,m]$, $\beta_i \geq 0, \forall j \in [1,...,s]$ and $\sum_{i=1}^{max(m,s)} (\alpha_i + \beta_i) < 1$.

$$
\\
$$

**parametric model**

\begin{align*}
r^M_t &= \phi_0 + \phi_1 r^M_{t-1} + ... + \phi_M r^M_{t-p} + \theta_1 \epsilon_{t-1} + ...+\theta_q\epsilon_{t-q} \; + \; \epsilon_t; \; \text{with} \; \epsilon_t \sim N (0,\sigma^2_{t-1}) \; \text{and} \\\\
\sigma^2_{t-1} &= \alpha_0 + \sum_{i=1}^m \alpha_i \epsilon^2_{t-i} \, + \, \sum_{j=1}^s \beta_j \sigma^2_{t-1-j}.
\end{align*}

$$
\\
$$

**joint log-likelihood:**

\begin{align*}
ln L(.) &= \sum_{t=1+max(p,q,m,s)}^T -\frac{1}{2} \ln (2\pi) - \frac{1}{2} \ln ( \sigma^2_{t-1} ) - \frac{1}{2} \, \frac{\epsilon^2_t}{ \sigma^2_{t-1}  } \quad \text{with} \; \\\\
\epsilon_t & \equiv r^M_t - \phi_0 - \phi_1 r^M_{t-1} - ... - \phi_p r^M_{t-p} - \theta_1 \epsilon_{t-1} - ... - \theta_q \epsilon_{t-q}, \quad \text{and} \\ \\
\sigma^2_{t-1} & \equiv  \alpha_0 + \sum_{i=1}^m \alpha_i \, \times \, \epsilon^2_{t-i} \, + \, \sum_{j=1}^s \beta_j \, \times \, \sigma^2_{t-1-j} \quad \text{for} \, t>1  \\ \\
& \text{and} \; \sigma^2_{0} \equiv \frac{\alpha_0}{ 1 -  \sum_{i=1}^m \alpha_i -  \sum_{j=1}^s \beta_j}.
\end{align*}


## B.2 Nonlinear Extensions for Skewness

 
**E-GARCH(1,1):**

\begin{align*}
& \ln (\sigma^2_{t-1}):= \alpha_0 +\alpha_{1} \times \eta_{t-1}+\theta_{1}(|\eta_{t-1}|-E[|\eta_{t-1}|])+\beta_{1} \times \ln(\sigma^2_{t-2})
\end{align*}

$$
\\
$$

with constant parameters $[\alpha_0, \alpha_1, \theta_1, \beta_1] \in \mathcal{R}^4$. 


**Re-write as:**
\begin{align*}
\ln  (\sigma^2_{t-1}) &:= \alpha_0 + g(\eta_{t-1}) + \beta_{1} \times \ln(\sigma^2_{t-2}), \\ \\
g(\eta_{t-1}) & :=    (\alpha_1 + \theta_1) \times I_{\eta_{t-1} > 0} \, \times \, \eta_{t-1} \, + \, (\alpha_1 - \theta_1) \times I_{\eta_{t-1} < 0} \, \times \, \eta_{t-1} \, - \, \theta_1 \times E[|\eta_{t-1}|].
\end{align*}

 
with

$$
E[|\eta_{t-1}|] = \sqrt{\frac{2}{\pi}}.
$$


$$
\\
$$

**The Threshold GARCH (1,1):**

\begin{align*}
\sigma^2_{t-1} & =\alpha_0 + ( \alpha_{1} + \theta_{1}\, \times \, I_{\epsilon_{t-1}<0)}) \, \times \, \epsilon_{t-1}^{2} + \beta_{1} \, \times \, \sigma^2_{t-2}.
\end{align*}

$$
\\
$$

 
-  Engle and Ng (1993) argue that the Threshold GARCH is preferred to the E-GARCH

## E.1 Joint Log Likelihood

Remember: assuming Gaussian innovations one ends up with

$$
L_T(\phi_0, \phi_1, \alpha_0, \alpha_1) = \prod_{t=2}^T \frac{1}{\sqrt{ 2 \pi (\alpha_0 + \alpha_1 \epsilon^2_{t-1})}} \times \exp\left( -\frac{(r_t - [\phi_0 + \phi_1 r_{t-1}])^2}{2 (\alpha_0 + \alpha_1 \epsilon^2_{t-1})} \right)
$$

Remember:
$$
\ln (L_T(.)) = \sum_{t=2}^T -\frac{1}{2} \ln(2\pi [\alpha_0 + \alpha_1 \epsilon^2_{t-1}]) - \frac{1}{2}  \frac{(r_t - [\phi_0 + \phi_1 r_{t-1}])^2}{2 (\alpha_0 + \alpha_1 \epsilon^2_{t-1})} 
$$

```python
# -ln( L_T(.) )
import numpy as np
 

def Neg_loglikelihood_ar1_arch1(parameters, r_t):   # Parameters is a list of model parameters, here: [\phi_0, phi_1, alpha_0, alpha_1   ]  
    phi_0   = parameters[0]
    phi_1   = parameters[1]
    alpha_0 = parameters[2]
    alpha_1 = parameters[3]

    means = phi_0 + phi_1 * r_t.iloc[:-1].values
    eps   = r_t.iloc[1:].values - means
    vars_  = alpha_0 + alpha_1 * eps[:-1]**2
       
    loglikeli = np.sum(-0.5 * np.log(2 * np.pi * vars_) - (r_t.iloc[2:].values - means[1:])**2 / (2 * vars_))

    return -loglikeli
```

## E.2  Start  Values

I read in data and 2pass optimal parameters from the last python tutorial.

```python
#start values for AR(1)-ARCH(1) parameters from last week's Python for Financial Data Science material
phi0_start = 0.0204
phi1_start = 0.9962
alpha0_start = 0.0004
alpha1_start = 0.3157

ar1_arch1_params_start = [phi0_start, phi1_start, alpha0_start, alpha1_start]
```

```python
# plotting helpers
def compute_model_impl_series(params, obs):
    phi_0 = params[0]
    phi_1 = params[1]
    alpha_0 = params[2]
    alpha_1 = params[3]
    ar1_impl_mean = phi_0 + phi_1 * obs.iloc[:-1].values
    eps = obs.iloc[1:].values - ar1_impl_mean
    arch_1_impl_var = alpha_0 + alpha_1 * eps[:-1] ** 2
    return ar1_impl_mean, arch_1_impl_var


def compute_ar1_forecast(params, horizon, obs_last):
    phi_0, phi_1, alpha_0, alpha_1 = params
    mean_next = obs_last
    for i in range(horizon):
        mean_next = phi_0 + phi_1 * mean_next
    return mean_next


def compute_arch1_forecast(params, horizon, var_last):
    phi_0, phi_1, alpha_0, alpha_1 = params
    var_next = var_last
    for i in range(horizon):
        var_next = alpha_0 + alpha_1 * var_next
    return var_next


def forecast_based_on_last_obs(params, obs_tail_1, obs_tail_2, horizons):
    phi_0, phi_1, alpha_0, alpha_1 = params
    ar1_impl_mean = phi_0 + phi_1 * obs_tail_2
    eps = obs_tail_1 - ar1_impl_mean
    ar1_forecasts = {}
    arch1_forecasts = {}
    for horizon in horizons:
        ar1_forecasts["h.{:02d}".format(horizon)] = compute_ar1_forecast(params=params, horizon=horizon,
                                                                         obs_last=obs_tail_1)
        arch1_forecasts["h.{:02d}".format(horizon)] = compute_arch1_forecast(params=params, horizon=horizon,
                                                                             var_last=eps ** 2)
    return ar1_forecasts, arch1_forecasts
```

```python
def interprete_optimized_params(params, obs):
    phi_0 = params[0]
    phi_1 = params[1]
    alpha_0 = params[2]
    alpha_1 = params[3]
    
    #print optimal AR(1)-ARCH(1) parameters 
    print("phi0: {}, phi1: {}, alpha0: {}, alpha1: {}".format(phi_0, phi_1, alpha_0, alpha_1))
    
    # print -1 * ln L_T
    llk = Neg_loglikelihood_ar1_arch1(params, obs)
    print("llk: {}".format(llk))
    
    d = {'phi_0': phi_0, 'phi_1': phi_1, 'alpha_0': alpha_0, 'alpha_1':alpha_1, 'neg_llk':llk}
    ser = pd.Series(data=d, index=['phi_0', 'phi_1', 'alpha_0', 'alpha_1', 'neg_llk'])
    
    return ser
```

```python
def compute_and_plot_model_impl_and_forecast(params, obs, axs, horizons=np.array(range(1, 13))):
    obs_yield_tail_1 = obs.tail(1).values[0]
    obs_yield_tail_2 = obs.tail(2).values[0]
    ar1_impl_mean, arch_1_impl_var = compute_model_impl_series(params=params, obs=y10_df)
    ar1_forecasts, arch1_forecasts = forecast_based_on_last_obs(params=params, obs_tail_1=obs_yield_tail_1,
                                                                obs_tail_2=obs_yield_tail_2,
                                                                horizons=horizons)
    axs[0].set_title("ar1_impl_mean")
    axs[0].plot(ar1_impl_mean)

    axs[1].set_title("arch_1_impl_var")
    axs[1].plot(arch_1_impl_var)

    axs[2].set_title("ar1_forecasts")
    axs[2].plot(pd.DataFrame.from_dict(ar1_forecasts, orient='index', columns=['ar1_forecasts']))

    axs[3].set_title("arch1_forecasts")
    axs[3].plot(pd.DataFrame.from_dict(arch1_forecasts, orient='index', columns=['arch1_forecasts']))
    return ar1_impl_mean, arch_1_impl_var, ar1_forecasts, arch1_forecasts
```

## E.3 Optimization 

### E.3.1 Nelder Mead

```python
#min (-ln L_T)
import scipy.optimize as sco

#Non-Constrained, non-linear optimization, i.e. https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html
ar1_arch1_params_optimal = sco.minimize(Neg_loglikelihood_ar1_arch1, ar1_arch1_params_start, args=(y10_df,),
                                        method='Nelder-Mead', options={'disp': True})

opt_params_nm = ar1_arch1_params_optimal.x
opt_params_llk_nm = interprete_optimized_params(params=opt_params_nm, obs=y10_df)
```

```python
opt_params_llk_nm =interprete_optimized_params(params=ar1_arch1_params_optimal.x, obs=y10_df)
```

```python
fig_nm, axs_nm = plt.subplots(4, 1)
fig_nm.suptitle("results of nm")
fig_nm.set_size_inches(12, 8)
_, _, arch1_forecasts_nm, ar1_forecasts_nm = compute_and_plot_model_impl_and_forecast(params=opt_params_nm, obs=y10_df,
                                                                                      axs=axs_nm)
plt.tight_layout()
plt.show()
```

**Observations:**

### E.3.2  L-BFGS-B

```python
lb_all = np.zeros(4)
ub_all = 1 * np.ones(4)
ub_all[1] = 0.999999
ub_all[3] = 0.999999
bnds_all = tuple((lb_all[x], ub_all[x]) for x in range(0, 4))
print(bnds_all)
```

```python
ar1_arch1_params_optimal_L_BFGS_B = sco.minimize(Neg_loglikelihood_ar1_arch1, ar1_arch1_params_start, args=(y10_df,), 
                                             method="L-BFGS-B", bounds=bnds_all,
                                            options={'ftol': 1e-12, 'gtol': 1e-12})
```

```python
opt_params_llk_lbfgsb = interprete_optimized_params(params=ar1_arch1_params_optimal_L_BFGS_B.x, obs=y10_df)
```

```python
opt_params_lbfgs = ar1_arch1_params_optimal_L_BFGS_B.x
fig_lbfgsb, axs_lbfgsb = plt.subplots(4, 1)
fig_lbfgsb.set_size_inches(12, 8)
fig_lbfgsb.suptitle("results of lbfgsb")
_, _, arch1_forecasts_lbfgsb, ar1_forecasts_lbfgsb = compute_and_plot_model_impl_and_forecast(params=opt_params_lbfgs,
                                                                                              obs=y10_df,
                                                                                              axs=axs_lbfgsb)
plt.tight_layout()
plt.show()
```
